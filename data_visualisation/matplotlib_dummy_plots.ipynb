```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
```

```python
# Params
a = - 2.5
b = 100.0
c = 0.0
eps = 1
lam = 10.0
mu = 0.0
sig = 100.0
n_points = 1000
lower_limit = -3.0
upper_limit = 3.0
data_range = upper_limit - lower_limit
```

```python
# Dummy scatter plot data
x = np.arange(lower_limit, upper_limit, data_range / n_points)
y1 = np.exp(- x ** 2) * np.random.normal(mu, sig, n_points)
y2 = a * x ** 2 + (b + eps * np.random.normal(mu, sig, n_points)) * x + c
z = 10 * np.exp(- x ** 2) * np.random.poisson(lam, n_points) 
```

```python
plt.scatter(x, y1, z, alpha = 0.2)
plt.scatter(x, y2, z, alpha = 0.2)
plt.xlabel('Dummy x variable')
plt.ylabel('Dummy y variable')
plt.title('Dummy scatter plot ')
plt.xlim(lower_limit, upper_limit)
plt.ylim(-500, 500)
plt.show()
```

```python
# Dummy boxplot data
y1 = np.random.normal(mu, sig, n_points)
y2 = np.random.normal(mu + 50, sig - 50, n_points)
data = [y1, y2]
```

```python
fig, ax = plt.subplots()
ax.set_title('Dummy boxplots')
ax.boxplot(data)

plt.xlabel('Dummy x variable ')
plt.ylabel('Dummy y variable')
plt.show()
```

```python
# Dummy time series data
y1 = np.empty(n_points)
y2 = np.empty(n_points)
mu = 0.0
sig = 0.01
y1[0] = 1.0
y2[0] = 1.0
for i in range(1, n_points):
    y1[i] = y1[i-1] * np.exp(np.random.normal(mu, sig))
    y2[i] = y2[i-1] * np.exp(np.random.normal(mu, sig))
```

```python
plt.plot(x, y1, linestyle = '-')
plt.plot(x, y2, linestyle = '-')
plt.title('Dummy time series line plot')
plt.xlabel('Time')
plt.ylabel('Dummy y variable')
plt.show()
```

```python
from scipy.stats import poisson

# Dummy PDF data
y1 = np.random.poisson(lam, 1000)
y2 = np.random.poisson(lam / 2, 1000)
y3 = np.random.poisson(lam / 4, 1000)
n_bins = 25
bins = np.arange(0, 25, 25 / n_bins)
```

```python
plt.hist(y1, bins, density = True, alpha = 0.5)
plt.hist(y2, bins, density = True, alpha = 0.5)
plt.hist(y3, bins, density = True, alpha = 0.5)
plt.title('Dummy histogram using Poisson-distibuted random variates')
plt.xlabel('Dummy x variable')
plt.ylabel('PDF')
plt.show()
```

```python
plt.hist(y1, bins, density = True, alpha = 0.5, cumulative = True)
plt.hist(y2, bins, density = True, alpha = 0.5, cumulative = True)
plt.hist(y3, bins, density = True, alpha = 0.5, cumulative = True)
plt.title('Dummy histogram using Poisson-distibuted random variates')
plt.xlabel('Dummy x variable')
plt.ylabel('PDF')
plt.show()
```
