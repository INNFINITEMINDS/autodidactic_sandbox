```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
# Exploring the data set.
df.head()
```

```python
df.info()
df.describe()
```

```python
# Correlations between numeric variables.
corr_mat = df.corr()
print(corr_mat)
sns.heatmap(corr_mat, square = True, center = 0)
plt.title('Correlation matrix heatmap')
plt.show()
```

```python
# Scatter plots and histograms between combinations of numeric variables.
sns.set_style('whitegrid')
g = sns.PairGrid(df, hue = 'species', diag_sharey = False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 2)
plt.show()
```

```python
# Melt into long format.
df_melted = df.melt(id_vars = 'species')
print(df_melted.head())
```

```python
g = sns.catplot(x="species", y="value",
            col="variable", aspect=.6,
            kind="violin", data=df_melted)
(g.set_axis_labels('', '')
.set_titles("{col_name}"))
plt.show()
```
