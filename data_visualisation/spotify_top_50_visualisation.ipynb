# Spotify top 50 songs 2019 - Data visualisation

## Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read in dataset

```python
top_50 = pd.read_csv('~/documents/Data/Spotify/top50_2019.csv', index_col = 0, encoding="ISO-8859-1")
```

## Extract numeric columns

```python
numeric = top_50.select_dtypes(include = 'int64')
```

## Plot

```python
sns.set_style('whitegrid')
g = sns.PairGrid(numeric, diag_sharey = False)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw = 2)
plt.show()
```
