```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import decomposition
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Import dataset and assign to X and y variables.
iris = load_iris()
X = pd.DataFrame(iris.data)
y = pd.DataFrame(data = {'Species': iris.target})

species_names = {0: 'Setosa',
                1: 'Versicola',
                2: 'Virginica'}

y = y.replace(to_replace = species_names)
```

```python
# Standardise the features.
X = StandardScaler().fit_transform(X)
```

```python
pca = PCA(n_components = 2)
principal_components = pca.fit_transform(X)
principal_components_df = pd.DataFrame(data = principal_components, columns = ['PC1', 'PC2'])

df_to_plot = pd.concat([principal_components_df, y], axis = 1)
```

```python
sns.scatterplot(x = 'PC1', y = 'PC2', hue = 'Species', data = df_to_plot)
plt.title('Plot to show first two principal components of iris dataset')
plt.show()
```
