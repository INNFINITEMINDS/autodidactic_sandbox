```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
X, y = load_iris(return_X_y=True)
print(X[0:10,])
print(y[0:10])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42)
```

```python
# Instantiate the model.
lin_reg = linear_model.LinearRegression()

# Fit the model to the dataset.
lin_reg.fit(X_train, y_train)

# Calculate the model score.
lin_reg_score = lin_reg.score(X_test, y_test)
print(lin_reg_score)
```

```python
# Define the parameter space.
alpha_space = np.arange(0.1, 1.1, 0.1)

# Initialise the cross-validation scores array.
cv_scores = []

# Fit the model to the dataset.
for a in alpha_space:
    ridge_reg = linear_model.Ridge(alpha = a, normalize = True)
    cv_score = np.mean(cross_val_score(ridge_reg, X_train, y_train, cv = 5))
    cv_scores.append(cv_score)

print(pd.DataFrame({'Alpha': alpha_space,
                   'Cross-validation scores': cv_scores}))

plt.plot(alpha_space, cv_scores, linewidth = 1)
plt.suptitle('Ridge regression on iris dataset', size = 14)
plt.title('Cross-validation score against alpha parameter')
plt.xlabel('Alpha parameter')
plt.ylabel('Cross-validation score')
plt.show()
```

```python
# Define the parameter space.
alpha_space = np.arange(0.1, 1.1, 0.1)

# Initialise the cross-validation scores array.
cv_scores = []

# Fit the model to the dataset.
for a in alpha_space:
    lasso_reg = linear_model.Lasso(alpha = a)
    cv_score = np.mean(cross_val_score(lasso_reg, X_train, y_train, cv = 5))
    cv_scores.append(cv_score)

print(pd.DataFrame({'Alpha': alpha_space,
                   'Cross-validation scores': cv_scores}))

plt.plot(alpha_space, cv_scores, linewidth = 1)
plt.suptitle('LASSO regression on iris dataset', size = 14)
plt.title('Cross-validation score against alpha parameter')
plt.xlabel('Alpha parameter')
plt.ylabel('Cross-validation score')
plt.show()
```

```python
# Define the parameter space.
alpha_space = np.arange(0.1, 1.1, 0.1)
l1_space = np.arange(0.1, 1.1, 0.1)

# Initialise the cross-validation scores array.
cv_scores = np.empty((10,10))

# Fit the model to the dataset.
for i in range(0, 10):
    for j in range(0, 10):
        elastic_net = linear_model.ElasticNet(alpha = alpha_space[i], l1_ratio = l1_space[j])
        cv_score = np.mean(cross_val_score(elastic_net, X_train, y_train, cv = 5))
        cv_scores[i, j] = cv_score

fig, ax = plt.subplots(1,1)
img = ax.imshow(cv_scores, extent=[0,1,0,1])
plt.suptitle('Elastic net model on iris dataset', size = 14)
plt.title('Cross-validation score against alpha parameter and l1-ratio parameter')
plt.xlabel('Alpha parameter')
plt.ylabel('l1-ratio')
fig.colorbar(img)
plt.show()
```
