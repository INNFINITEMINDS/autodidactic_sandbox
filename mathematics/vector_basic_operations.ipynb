```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([5, -7, 4, 2, 3, 5])
B = np.array([1, 2, 4, 9, 7, 3])

plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
# Vector addition
D = A + B
```

```python
plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue')
plt.arrow(A[0], A[1], B[0], B[1], color = 'blue')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue')
plt.arrow(A[0], A[1], B[0], B[1], color = 'blue')
plt.arrow(0, 0, D[0], D[1], color = 'purple')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue')
plt.arrow(A[0], A[1], B[0], B[1], color = 'blue')
plt.arrow(B[0], B[1], A[0], A[1], color = 'red')
plt.arrow(0, 0, D[0], D[1], color = 'purple')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
# Scalar multiplication
c = 3
D = c * B

plt.arrow(0, 0, D[0], D[1], color = 'green')
plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue', linewidth = 3)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```

```python
# Linear combination
d = 0.5
E = c * B + d * A

plt.arrow(0, 0, A[0], A[1], color = 'red')
plt.arrow(0, 0, B[0], B[1], color = 'blue')
plt.arrow(0, 0, E[0], E[1], color = 'black')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid()
plt.show()
```
